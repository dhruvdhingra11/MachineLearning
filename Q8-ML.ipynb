# Create various type of plots/charts like histograms, plot based on sine/cosine function based on data from a matrix. Further label different axes in a plot and data in a plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
matrix = np.array([1,2,3,4,5,6,1,1,1,2,2,2,3,3,3,7,4])
```

```python
# plotting a histogram
plt.hist(matrix)
plt.show()
```

```python
# plotting pie chart
arr = np.array([1,2,3,4])
labelss = ['Dhruv','Naman','debarpann','cc']
plt.pie(arr,labels=labelss)
plt.legend()
plt.title('Pie Chart')
plt.show()
```

```python
# Plotting bar chart
arr3 = ['A','B','C','D','E','F']
arr2 = [8,9,10,47,11,9]
plt.bar(arr3,arr2,color='red')
plt.legend()
plt.title('Bar chart')
plt.show()
```

# plotting sine cosine

```python
# plotting sine cosine
x = np.arange(0,10,0.5)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y,color='green',label='sin')
plt.plot(x,z,color='red',label='cos')
plt.show()
```
