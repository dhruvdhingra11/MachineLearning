```python
import matplotlib.pyplot as plt
import numpy as np


age_g1 = [1, 3, 5, 10, 15, 17, 18, 16, 19, 21, 23, 28, 30, 31, 33, 38, 32, 40, 45, 43, 49, 55, 53, 63, 66, 85, 80, 57, 75, 93, 95]
age_g2 = [6, 4, 15, 17, 19, 21, 28, 23, 31, 36, 39, 32, 50, 56, 59, 74, 79, 34, 98, 97, 95, 67, 69, 92, 45, 55, 77, 76, 85]


fig, ax = plt.subplots()


ax.hist([age_g1, age_g2], bins=10, label=['Age Group 1', 'Age Group 2'], edgecolor='black', alpha=0.7)


ax.set_xlabel('Age')
ax.set_ylabel('Count')
ax.set_title('Histogram of Age Groups')
ax.legend()


plt.show()


fig, ax = plt.subplots()


ax.hist([age_g1, age_g2], bins=14, label=['Age Group 1', 'Age Group 2'], edgecolor='black', alpha=0.7)


ax.set_xlabel('Age')
ax.set_ylabel('Count')
ax.set_title('Histogram of Age Groups')
ax.legend()


plt.show()
```

```python

IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]


width = 0.2
plt.bar(np.arange(2015, 2020)-width, IT, width=width, label='IT')
plt.bar(np.arange(2015, 2020), ECE, width=width, label='ECE')
plt.bar(np.arange(2015, 2020)+width, CSE, width=width, label='CSE')


plt.xlabel('Year')
plt.ylabel('Number of Students Passed')
plt.title('Engineering Students Passed by Branch (2015-2019)')


plt.legend()


plt.show()
```

```python
men_means = (22, 30, 35, 35, 26)
women_means = (25, 32, 30, 35, 29)
men_std = (4, 3, 4, 1, 5)
women_std = (3, 5, 2, 3, 3)

ind = np.arange(len(men_means))    
width = 0.35                      

fig, ax = plt.subplots()


rects1 = ax.bar(ind, men_means, width, yerr=men_std, color='b')


rects2 = ax.bar(ind, women_means, width, yerr=women_std, bottom=men_means, color='r')


ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))


ax.legend((rects1[0], rects2[0]), ('Men', 'Women'))

plt.show()
```

```python
year = [2014, 2015, 2016, 2017, 2018, 2019]
issues_addressed = [10, 14, 0, 10, 15, 15]
issues_pending = [5, 10, 50, 2, 0, 10]


fig, ax = plt.subplots()

ax.set_xlim(0, sum(issues_addressed) + sum(issues_pending))

ax.barh(year, issues_addressed, color='green')
ax.barh(year, issues_pending, left=issues_addressed, color='red')


ax.set_xlabel('Number of issues')
ax.set_ylabel('Year')
ax.set_title('Issues Addressed and Pending by Year')
ax.legend(['Addressed', 'Pending'])


plt.show()
```
